(e) XGBoost and Model Trees

```python
x_train=train.iloc[:, 1:]
y_train=train.iloc[:, :1]
x_test=test.iloc[:, 1:]
y_test=test.iloc[:, :1]
```

```python
y_train=LabelEncoder().fit_transform(y_train.values.ravel())
y_test=LabelEncoder().fit_transform(y_test.values.ravel())
```

```python
alphas=[10**-5,10**-4,10**-3,10**-2,10**-1,10**0,10**1,10**2,10**3,10**4,10**5]
alphas
```

**Ref:Chatgpt how to use crossvalidation using XGB classifier**

```python
accuracy_for_alps={}
for i in alphas:
    xgb_model=XGBClassifier(objective='binary:logistic',reg_alpha=i,random_state=42)
    score=cross_val_score(xgb_model,x_train,y_train,cv=10,scoring='accuracy')
    accuracy_for_alps[str(i)]=score.mean()
max_alpha=max(accuracy_for_alps,key=accuracy_for_alps.get)
print("alpha value selected is ",max_alpha)
```

```python
max_alpha = max(accuracy_for_alps, key=accuracy_for_alps.get)
print("alpha value selected is ",max_alpha)
```

```python
alpha_val=float(max_alpha)
xgb_model=XGBClassifier(objective='binary:logistic',reg_lambda=alpha_val,random_state=42)
xgb_model.fit(x_train,y_train)
score=cross_val_score(xgb_model,x_train,y_train,cv=10,n_jobs=-1)
acc=score.mean()
print("The training error using 10 fold CV using the selected alpha is :",1-acc)
```

```python
score= cross_val_score(xgb_model,x_test,y_test,cv=10,n_jobs=-1)
acc=score.mean()
print("The testing error using 10 fold CV using the selected alpha is :",1-acc)
```

```python
xgb_model.fit(x_train, y_train)
xgb_y_train_pred=xgb_model.predict(x_train)
xgb_y_test_pred=xgb_model.predict(x_test)
```

```python
xgb_conf_matrix=confusion_matrix(y_train,xgb_y_train_pred)
xgb_conf_matrix
```

```python
RocCurveDisplay.from_estimator(xgb_model,x_train,y_train)
plt.title("ROC/AUC on Training Set")
plt.show()
```

```python
xgb_conf_matrix_test=confusion_matrix(y_test,xgb_y_test_pred)
xgb_conf_matrix_test
```

```python
RocCurveDisplay.from_estimator(xgb_model,x_test,y_test)
plt.title("ROC/AUC on Testing Set")
plt.show()
```

(f)Use SMOTE (Synthetic Minority Over-sampling Technique)

**Reg:Chatgpt how to combine SMOTE and XGB classifier within cv**

```python
pipeline=imbpipeline([['smote', SMOTE(random_state=42)],['xgb_model',XGBClassifier(random_state=42,verbosity=0)]])
param_grid={'xgb_model__reg_alpha':alphas}
grid_search=GridSearchCV(estimator=pipeline,param_grid=param_grid,cv=10,refit=True)
grid_search.fit(x_train, y_train)
```

```python
score= cross_val_score(grid_search,x_train,y_train,cv=10,n_jobs=-1)
acc=score.mean()
print("(SMOTE)The training error using 10 fold CV is :",1-acc)
```

```python
score= cross_val_score(grid_search, x_test, y_test, cv=10, n_jobs=-1)
acc=score.mean()
print("(SMOTE)The testing error using 10 fold CV is :",1-acc)
```

```python
grid_search.fit(x_train, y_train)
smote_y_train_pred = grid_search.predict(x_train)
smote_y_test_pred = grid_search.predict(x_test)
```

```python
smote_conf_matrix=confusion_matrix(y_train,smote_y_train_pred)
smote_conf_matrix
```

```python
RocCurveDisplay.from_estimator(grid_search,x_train,y_train)
plt.title("(SMOTE)ROC/AUC on Training Set")
plt.show()
```

```python
RocCurveDisplay.from_estimator(grid_search,x_test,y_test)
plt.title("(SMOTE)ROC/AUC on Testing Set")
plt.show()
```

The uncompensated case performs better than SMOTE case.

(b)Test RSS- Decrease initially, and then eventually start increasing in a U shape. As s increases, constraint on beta decreases, Test RSS reduces and eventually starts increasing as beta starts overfitting. 

(c)Variance-Steadily increase.As s increases, constraint on beta decreases which will lead to more flexibilty leading to higher variance 

(d)(Squared) bias-Steadily decrease.As s increases, constraint on beta decreases which will lead to more flexibilty and fitting the data better and reducing the bias

(e)Irreducible error-Remains constant. It is not dependant on the features of the model like s
