```python
path_to_data = "../data/RealWaste"
train_dir = "../data/train"
test_dir = "../data/test"
img_size = 128  
split_ratio = 0.8 

class_names = sorted([d for d in os.listdir(path_to_data) if os.path.isdir(os.path.join(path_to_data, d))])
num_classes = len(class_names)


os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Loop through each class and split images based on the first 80% for training
for class_name in class_names:

    class_path = os.path.join(path_to_data, class_name)
    train_class_path = os.path.join(train_dir, class_name)
    test_class_path = os.path.join(test_dir, class_name)
    

    os.makedirs(train_class_path, exist_ok=True)
    os.makedirs(test_class_path, exist_ok=True)


    image_files = sorted(os.listdir(class_path))

  
    split_index = int(len(image_files) * split_ratio)


    train_files = image_files[:split_index]
    test_files = image_files[split_index:]

 
    for file in train_files:
        shutil.copy(os.path.join(class_path, file), os.path.join(train_class_path, file))
    for file in test_files:
        shutil.copy(os.path.join(class_path, file), os.path.join(test_class_path, file))

print("Data split completed! Training and testing directories are ready.")
```

```python
train_original_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)
train_aug_datagen = ImageDataGenerator(
    rescale=1./255,               
    rotation_range=5,             
    zoom_range=0.1,                
    width_shift_range=0.1,         
    height_shift_range=0.1,        
    horizontal_flip=True,         
    brightness_range=[0.7, 1.3],  
    validation_split=0.2
)
train_original_generator = train_original_datagen.flow_from_directory(
    train_dir,                     
    target_size=(img_size, img_size),  
    batch_size=5,                
    class_mode='categorical',
    subset='training', 
    shuffle=True                   
)
val_original_generator = train_original_datagen.flow_from_directory(
    train_dir,                     
    target_size=(img_size, img_size),  
    batch_size=5,                
    class_mode='categorical',
    subset='validation', 
    shuffle=True                   
)
train_aug_generator = train_aug_datagen.flow_from_directory(
    train_dir,                     
    target_size=(img_size, img_size),  
    batch_size=5,                 
    class_mode='categorical',
    subset='training', 
    shuffle=True                  
)

def doubled_data_generator(original_gen, aug_gen):
    """
    Yields batches that include both original and augmented images.
    For each epoch, you’ll go through 2× the original data.
    """
    while True:
        
        x_orig, y_orig = next(original_gen)
        x_aug, y_aug = next(aug_gen)

       
        x_combined = np.concatenate((x_orig, x_aug), axis=0)
        y_combined = np.concatenate((y_orig, y_aug), axis=0)

        
        indices = np.random.permutation(len(x_combined))
        x_shuffled = x_combined[indices]
        y_shuffled = y_combined[indices]

        
        batch_size = len(x_orig)
        for i in range(0, len(x_shuffled), batch_size):
            yield x_shuffled[i:i+batch_size], y_shuffled[i:i+batch_size]

train_generator = doubled_data_generator(train_original_generator, train_aug_generator)

test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir,                     
    target_size=(img_size, img_size),  
    batch_size=5,                 
    class_mode='categorical',      
    shuffle=False                  
)

early_stopping = EarlyStopping(
    monitor='val_loss', 
    patience=50,  
    restore_best_weights=True,  
    verbose=1
)
checkpoint = ModelCheckpoint(
    filepath='best_model.h5',        
    monitor='val_loss',              
    save_best_only=True,             
    save_weights_only=False,         
    verbose=1
)



```

```python
def plot_hist(hist):
    plt.figure(figsize=(8, 6), dpi=80)
    plt.plot([1 - i for i in hist.history["accuracy"]])
    plt.plot([1 - i for i in hist.history["val_accuracy"]])
    plt.title("Training and Validation Errors vs. Epochs")
    plt.ylabel("Error")
    plt.xlabel("Epochs")
    plt.legend(["train", "validation"], loc="upper right")
    plt.show()
```

**Data augmentation and train, validation and test set generation**

```python
train_original_datagen_trial = ImageDataGenerator(
    rescale=1,
    validation_split=0.2
)
train_aug_datagen_trial = ImageDataGenerator(
    rescale=1,                
    rotation_range=5,             
    zoom_range=0.1,                
    width_shift_range=0.1,         
    height_shift_range=0.1,        
    horizontal_flip=True,          
    brightness_range=[0.7, 1.3],   
    validation_split=0.2
)
train_original_generator_trial = train_original_datagen_trial.flow_from_directory(
    train_dir,                     
    target_size=(img_size, img_size),  
    batch_size=5,                 
    class_mode='categorical',
    subset='training'                 
)
val_original_generator_trial = train_original_datagen_trial.flow_from_directory(
    train_dir,                     
    target_size=(img_size, img_size),  
    batch_size=5,                 
    class_mode='categorical',
    subset='validation'                  
)
train_aug_generator_trial = train_aug_datagen_trial.flow_from_directory(
    train_dir,                     
    target_size=(img_size, img_size),  
    batch_size=5,                 
    class_mode='categorical',
    subset='training'                 
)
train_generator_trial = doubled_data_generator(train_original_generator_trial, train_aug_generator_trial)
test_datagen_trial = ImageDataGenerator(rescale=1)
test_generator_trial = test_datagen_trial.flow_from_directory(
    test_dir,                      
    target_size=(img_size, img_size),  
    batch_size=5,                
    class_mode='categorical'                
)
```

**EfficientNetB0**

```python
checkpoint_3 = ModelCheckpoint(
    filepath='best_model_three.keras',        
    monitor='val_loss',              
    save_best_only=True,             
    save_weights_only=False,         
    verbose=1
)
base_model_three = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(img_size, img_size, 3))
base_model_three.trainable = False
model_three = models.Sequential([
    base_model_three,
    GlobalAveragePooling2D(),
    BatchNormalization(),
    Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.2),
    BatchNormalization(),
    Dense(num_classes, activation='softmax')
])
model_three.compile(
        optimizer=Adam(learning_rate=1e-4),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
steps_per_epoch = (train_original_generator_trial.samples * 2) // train_original_generator_trial.batch_size
history_three = model_three.fit(
        train_generator_trial,
        steps_per_epoch=steps_per_epoch,
        epochs=100,
        validation_data=val_original_generator_trial,
        callbacks=[checkpoint_3, early_stopping]
    )
```
