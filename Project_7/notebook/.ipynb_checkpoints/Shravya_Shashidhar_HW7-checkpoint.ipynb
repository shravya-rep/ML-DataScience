2. K-Means Clustering on a Multi-Class and Multi-Label Data Set

(a)Use k-means clustering on the whole Anuran Calls (MFCCs) Data Set (do not split
the data into train and test, as we are not performing supervised learning in this
exercise). Choose k 2 f1; 2; : : : ; 50g automatically based on one of the methods
provided in the slides (CH or Gap Statistics or scree plots or Silhouettes) or any
other method you know.

```python
data=pd.read_csv("../data/Anuran Calls (MFCCs)/Frogs_MFCCs.csv")
data_for_kmeans=data.drop(columns=['Species', 'Genus', 'Family', 'RecordID'], errors='ignore')
data_for_kmeans
```

```python
silhouette_scores=[]
k_values=range(2, 51)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(data_for_kmeans)
    score = silhouette_score(data_for_kmeans, labels)
    silhouette_scores.append(score)
best_k = k_values[np.argmax(silhouette_scores)]
print("The choosen k is",best_k)
```

```python
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(data_for_kmeans)
data['Cluster'] = cluster_labels
data
```

Ref:chatgpt "find majority among clusters after kmeans"

```python
cluster_majority_labels = {}
for cluster_id in sorted(data['Cluster'].unique()):
    cluster_data = data[data['Cluster'] == cluster_id]
    family_majority = cluster_data['Family'].value_counts().idxmax()
    genus_majority = cluster_data['Genus'].value_counts().idxmax()
    species_majority = cluster_data['Species'].value_counts().idxmax()
    cluster_majority_labels[cluster_id] = (family_majority, genus_majority, species_majority)
    
    print(f"\nCluster {cluster_id}:")
    print(f"  Most common Family:  {family_majority}")
    print(f"  Most common Genus:   {genus_majority}")
    print(f"  Most common Species: {species_majority}")
```

```python
data['Pred_Family'] = data['Cluster'].map(lambda x: cluster_majority_labels[x][0])
data['Pred_Genus'] = data['Cluster'].map(lambda x: cluster_majority_labels[x][1])
data['Pred_Species'] = data['Cluster'].map(lambda x: cluster_majority_labels[x][2])
```

```python
true_labels = data[['Family', 'Genus', 'Species']].values
pred_labels = data[['Pred_Family', 'Pred_Genus', 'Pred_Species']].values
hamming_distances = np.sum(true_labels != pred_labels, axis=1)
average_hamming_distance = np.mean(hamming_distances)
print("The average hamming distance is",average_hamming_distance)
incorrect_labels = np.sum(true_labels != pred_labels, axis=1)
average_hamming_loss = np.sum(incorrect_labels) / (len(data) * 3)
print("The average hamming loss is",average_hamming_loss)
correct_labels = np.sum(true_labels == pred_labels, axis=1)
average_hamming_score = np.sum(correct_labels) / (len(data) * 3)
print("The average hamming score is",average_hamming_score)
```

```python
num_iterations = 50
hamming_distances = []
for i in range(num_iterations):
    print("Complete")
    silhouette_scores=[]
    k_values=range(2, 51)
    for k in k_values:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(data_for_kmeans)
        score = silhouette_score(data_for_kmeans, labels)
        silhouette_scores.append(score)
    best_k = k_values[np.argmax(silhouette_scores)]

    kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(data_for_kmeans)
    data['Cluster'] = cluster_labels

    cluster_majority_labels = {}
    for cluster_id in sorted(data['Cluster'].unique()):
        cluster_data = data[data['Cluster'] == cluster_id]
        family_majority = cluster_data['Family'].value_counts().idxmax()
        genus_majority = cluster_data['Genus'].value_counts().idxmax()
        species_majority = cluster_data['Species'].value_counts().idxmax()
        cluster_majority_labels[cluster_id] = (family_majority, genus_majority, species_majority)

    data['Pred_Family'] = data['Cluster'].map(lambda x: cluster_majority_labels[x][0])
    data['Pred_Genus'] = data['Cluster'].map(lambda x: cluster_majority_labels[x][1])
    data['Pred_Species'] = data['Cluster'].map(lambda x: cluster_majority_labels[x][2])
    true_labels = data[['Family', 'Genus', 'Species']].values
    pred_labels = data[['Pred_Family', 'Pred_Genus', 'Pred_Species']].values
    incorrect_labels = np.sum(true_labels != pred_labels, axis=1)
    average_hamming_distance = np.sum(incorrect_labels) / (len(data) * 3)
    hamming_distances.append(average_hamming_distance)

hamming_distances = np.array(hamming_distances)
average_distance = np.mean(hamming_distances)
std_distance = np.std(hamming_distances)
    
```
