Select the first 70 rows of Class 0 and the first 140 rows of Class 1 as the training set and the rest of the data as the test set.

```python
x_train = trainData[['pelvic_incidence', 'pelvic_tilt', 'lumbar_lordosis_angle', 
                     'sacral_slope', 'pelvic_radius', 'degree_spondylolisthesis']]
y_train = trainData[['class']]
```

```python
x_test = testData[['pelvic_incidence', 'pelvic_tilt', 'lumbar_lordosis_angle', 
                     'sacral_slope', 'pelvic_radius', 'degree_spondylolisthesis']]
y_test = testData[['class']]
```

Classification using KNN on Vertebral Column Data Set

i)Write code for k-nearest neighbors with Euclidean metric (or use a software package).

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, f1_score
```

Test all the data in the test database with k nearest neighbors. Take de-
cisions by majority polling. Plot train and test errors in terms of k for
k 2 f208; 205; : : : ; 7; 4; 1; g (in reverse order). You are welcome to use smaller
increments of k. Which k is the most suitable k among those values? Cal-
culate the confusion matrix, true positive rate, true negative rate, precision,
and F1-score when k = k.2

```python
trainErrors = []
testErrors = []

allK = [i for i in range(208, 0, -3)]
for k in allK:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train.values.ravel())
    y_train_pred = knn.predict(x_train)
    y_test_pred = knn.predict(x_test)
    train_error = 1 - accuracy_score(y_train, y_train_pred)
    test_error = 1 - accuracy_score(y_test, y_test_pred)
    trainErrors.append(train_error)
    testErrors.append(test_error)

plt.plot(allK, trainErrors, label='Train errors')
plt.plot(allK, testErrors, label='Test errors')
plt.xlabel('$k$')
plt.ylabel('Errors')
plt.legend(loc="upper right")
plt.gca().invert_xaxis()
plt.show()
    
```

```python
for i in range(0, len(trainErrors)):
    if trainErrors[i] == min(trainErrors):
        print('The optimal k* from training error data or the given data is', allK[i])
```

```python
for i in range(0, len(testErrors)):
    if testErrors[i] == min(testErrors):
        print('The optimal k* from testing error data for the given data is', allK[i])
```

Calculate the confusion matrix, true positive rate, true negative rate, precision,and F1-score when k = k* 

```python
knn4 = KNeighborsClassifier(n_neighbors=4)
knn4.fit(x_train, y_train.values.ravel())
yFinalPred = knn4.predict(x_test)
confMatrix = confusion_matrix(y_test, yFinalPred)
print('Confusion Matrix is \n', confMatrix)
```

```python
trueNegative=confMatrix[0][0]
falsePositive=confMatrix[0][1]
falseNegative=confMatrix[1][0]
truePositive=confMatrix[1][1]

truePositiveRate=truePositive/(falseNegative+truePositive)
trueNegativeRate=trueNegative/(trueNegative+falsePositive)

print('True Positive Rate is:', truePositiveRate)
print('True Negative Rate is:', trueNegativeRate)

```

```python
precision = precision_score(y_test, yFinalPred)
print('Precision score is', precision)
```

```python
f1Score= f1_score(y_test, yFinalPred)
print('F1 score is', f1Score)
```

Since the computation time depends on the size of the training set, one may
only use a subset of the training set. Plot the best test error rate,3 which
is obtained by some value of k, against the size of training set, when the
size of training set is N 2 f10; 20; 30; : : : ; 210g.4 Note: for each N, select
your training set by choosing the first bN=3c rows of Class 0 and the first
N 􀀀bN=3c rows of Class 1 in the training set you created in 1(b)iii. Also, for
each N, select the optimal k from a set starting from k = 1, increasing by 5.
For example, if N = 200, the optimal k is selected from f1; 6; 11; : : : ; 196g.
This plot is called a Learning Curve.

```python
optimalK=dict()
allN = [i for i in range(10, 210, 10)]

for n in allN:
    index1=math.floor(n/3)
    index2=n-math.floor(n/3)
    small_trainDataFor0 = trainDataFor0[0:index1]
    small_trainDataFor1 = trainDataFor1[0:index2]
    small_trainData=pd.concat([small_trainDataFor0,small_trainDataFor1])
    small_x_train = small_trainData[['pelvic_incidence', 'pelvic_tilt', 'lumbar_lordosis_angle', 
                     'sacral_slope', 'pelvic_radius', 'degree_spondylolisthesis']]
    small_y_train = small_trainData[['class']]
    small_testErrors = []
    kRange = [j for j in range(1, n, 5)]
    for k in kRange:
        small_knn = KNeighborsClassifier(n_neighbors=k)
        small_knn.fit(small_x_train,small_y_train.values.ravel())
        small_yFinalPred = small_knn.predict(x_test)
        small_test_error = 1 - accuracy_score(y_test, small_yFinalPred)
        small_testErrors.append(small_test_error)

    optimalK[n]=min(small_testErrors)

plt.plot(optimalK.keys(), optimalK.values())
plt.xlabel('N')
plt.ylabel('best test error rate')
        

```

i. Minkowski Distance:
A. which becomes Manhattan Distance with p = 1.

```python
newTestErrors1=[]
newTrainErrors1=[]
allTestErrors=dict()
kNewRange= [i for i in range(1, 196, 5)]

for k in kNewRange:
    knn_manhattan= KNeighborsClassifier(n_neighbors=k, p=1)
    knn_manhattan.fit(x_train, y_train.values.ravel())
    new_y_train_pred = knn_manhattan.predict(x_train)
    new_y_test_pred = knn_manhattan.predict(x_test)
    new_train_error= 1 - accuracy_score(y_train, new_y_train_pred)
    new_test_error = 1 - accuracy_score(y_test, new_y_test_pred)
    newTestErrors1.append(new_test_error)
    newTrainErrors1.append(new_train_error)

for i in range(0, len(newTestErrors1)):
    if newTestErrors1[i] == min(newTestErrors1):
        print('Test error =', newTestErrors1[i],'for k =', kNewRange[i])

allTestErrors['ManhattanDistance']=min(newTestErrors1)
```

```python
for i in range(0, len(newTrainErrors1)):
    if newTrainErrors1[i] == min(newTrainErrors1):
        print('Train error =', newTrainErrors1[i],'for k =', kNewRange[i])
```

K=1. leads to the case of overfitting, the next possible k=6 would be ideal value as k*

B. with log10(p) In this case, use the k* you found for the Manhattan distance in 1(d)iA. What is the best log10(p)?

```python
newTestErrors2=[]
newTrainErrors2=[]
pRange = [10**i for i in np.linspace(0.1, 1.0, 10)]

for p in pRange:
    knn_log= KNeighborsClassifier(n_neighbors=6, p=p)
    knn_log.fit(x_train, y_train.values.ravel())
    new_y_train_pred= knn_log.predict(x_train)
    new_y_test_pred = knn_log.predict(x_test)
    new_train_error = 1 - accuracy_score(y_train, new_y_train_pred)
    new_test_error = 1 - accuracy_score(y_test, new_y_test_pred)
    newTestErrors2.append(new_test_error)
    newTrainErrors2.append(new_train_error)

for i in range(0, len(newTestErrors2)):
    if newTestErrors2[i] == min(newTestErrors2):
        print('Test error =', newTestErrors2[i],'for p=', pRange[i], 'best 10log(p)=',math.log10(pRange[i]))

allTestErrors['Log MD']=min(newTestErrors2)
    
```

```python
for i in range(0, len(newTrainErrors2)):
    if newTrainErrors2[i] == min(newTrainErrors2):
        print('Train error =', newTrainErrors2[i],'for p=', pRange[i], 'best 10log(p)=',math.log10(pRange[i]))
```

C. which becomes Chebyshev Distance with p-> infinity

```python
newTestErrors3=[]
newTrainErrors3=[]
kNewRange= [i for i in range(1, 196, 5)]

for k in kNewRange:
    knn_chebyshev = KNeighborsClassifier(n_neighbors=k, p=math.inf)
    knn_chebyshev.fit(x_train, y_train.values.ravel())
    new_y_train_pred = knn_log.predict(x_train)
    new_y_test_pred = knn_chebyshev.predict(x_test)
    new_train_error = 1 - accuracy_score(y_train, new_y_train_pred)
    new_test_error = 1 - accuracy_score(y_test, new_y_test_pred)
    newTrainErrors3.append(new_train_error)
    newTestErrors3.append(new_test_error)

for i in range(0, len(newTestErrors3)):
    if newTestErrors3[i] == min(newTestErrors3):
        print('Test error =', newTestErrors3[i],'for k =', kNewRange[i])

allTestErrors['Chebyshev Distance']=min(newTestErrors3)
```

```python
for i in range(0, len(newTrainErrors3)):
    if newTrainErrors3[i] == min(newTrainErrors3):
        print('Train error =', newTrainErrors3[i],'for k =', kNewRange[i])
```

```python
ii. Mahalanobis Distance
```

```python
newTestErrors4=[]
newTrainErrors4=[]
kNewRange= [i for i in range(1, 196, 5)]

covMatrix= np.cov(x_train.T)
covMatrix_inv= np.linalg.pinv(covMatrix)

for k in kNewRange:
    knn_mahalanobis = KNeighborsClassifier(n_neighbors=k, metric='mahalanobis', metric_params={'VI': covMatrix_inv})
    knn_mahalanobis.fit(x_train, y_train.values.ravel())
    new_y_train_pred = knn_mahalanobis.predict(x_train)
    new_y_test_pred = knn_mahalanobis.predict(x_test)
    new_train_error= 1 - accuracy_score(y_train, new_y_train_pred)
    new_test_error = 1 - accuracy_score(y_test, new_y_test_pred)
    newTestErrors4.append(new_test_error)
    newTrainErrors4.append(new_train_error)

for i in range(0, len(newTestErrors4)):
    if newTestErrors4[i] == min(newTestErrors4):
        print('Test error is min =', newTestErrors4[i],'at k =', kNewRange[i])

allTestErrors['Mahalanobis Distance']=min(newTestErrors4)
    
```

```python
for i in range(0, len(newTrainErrors4)):
    if newTrainErrors4[i] == min(newTrainErrors4):
        print('Train error is min =', newTrainErrors4[i],'at k =', kNewRange[i])
```

(e) The majority polling decision can be replaced by weighted decision, in which the
weight of each point in voting is inversely proportional to its distance from the
query/test data point. In this case, closer neighbors of a query point will have
a greater influence than neighbors which are further away. Use weighted voting
with Euclidean, Manhattan, and Chebyshev distances and report the best test
errors when k={1; 6; 11; 16; : : : ; 196}

```python
WeightedTestErrors1=[]
WeightedTrainErrors1=[]
kNewRange= [i for i in range(1, 196, 5)]

for k in kNewRange:
    knn_norm = KNeighborsClassifier(n_neighbors=k,weights='distance')
    knn_norm.fit(x_train, y_train.values.ravel())
    new_y_train_pred = knn_norm.predict(x_train)
    new_y_test_pred = knn_norm.predict(x_test)
    new_train_error = 1 - accuracy_score(y_train, new_y_train_pred)
    new_test_error = 1 - accuracy_score(y_test, new_y_test_pred)
    WeightedTestErrors1.append(new_test_error)
    WeightedTrainErrors1.append(new_train_error)

for i in range(0, len(WeightedTestErrors1)):
    if WeightedTestErrors1[i] == min(WeightedTestErrors1):
        print('Min test error is ', WeightedTestErrors1[i],'for k =', kNewRange[i])
    
```

```python
for i in range(0, len(WeightedTrainErrors1)):
    if WeightedTrainErrors1[i] == min(WeightedTrainErrors1):
        print('Min train error is ', WeightedTrainErrors1[i],'for k =', kNewRange[i])
    
```
